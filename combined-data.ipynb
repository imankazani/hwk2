```python
import pandas as pd
from pathlib import Path
```

```python
data_2014 = pd.read_csv("../data/output/data-2014.csv")
data_2015 = pd.read_csv("../data/output/data-2015.csv")
data_2016 = pd.read_csv("../data/output/data-2016.csv")
data_2017 = pd.read_csv("../data/output/data-2017.csv")
data_2018 = pd.read_csv("../data/output/data-2018.csv")
data_2019 = pd.read_csv("../data/output/data-2019.csv")
```

```python
for name, df in [
    ("2014", data_2014),
    ("2015", data_2015),
    ("2016", data_2016),
    ("2017", data_2017),
    ("2018", data_2018),
    ("2019", data_2019),
]:
    print(name, len(df.columns))
```

```python
print(set(data_2014.columns) - set(data_2019.columns))
print(set(data_2019.columns) - set(data_2014.columns))
```

```python
all_years = [
    data_2014,
    data_2015,
    data_2016,
    data_2017,
    data_2018,
    data_2019
]

master_data = pd.concat(all_years, ignore_index=True)
```

```python
print(master_data.shape)
print(master_data['year'].value_counts().sort_index())
```

```python
output_path = "../data/output/master_data.csv"
Path(output_path).parent.mkdir(parents=True, exist_ok=True)
master_data.to_csv(output_path, index=False)
print(f"Data saved to {output_path}")
```
